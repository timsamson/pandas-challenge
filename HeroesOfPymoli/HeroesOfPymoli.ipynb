```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)

purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data_df["SN"].unique())
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate Values
total_purchases = purchase_data_df["Price"].count()

total_revenue = purchase_data_df["Price"].sum()

unique_items = len(purchase_data_df["Item ID"].unique())

avg_price = total_revenue / total_purchases

#Create Data Frame
summary_df= pd.DataFrame(columns=['Total Purchases', 'Total Revenue', 'Unique Items', 'Average Price'])
summary_df = summary_df.append({'Total Purchases':total_purchases, 'Total Revenue':total_revenue, 
                                'Unique Items': unique_items, 'Average Price':avg_price}, ignore_index=True)


#Format and Clean Data
summary_df["Average Price"] = summary_df["Average Price"].astype(float).map("${:,.2f}".format)
summary_df["Total Purchases"] = summary_df["Total Purchases"].astype(int).map("{:}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
summary_df["Unique Items"] = summary_df["Unique Items"].astype(int).map("{:}".format)
summary_df

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Groupby Gender
gender_df = purchase_data_df.groupby(["Gender"])

#Total # of entries in gender
total_gender = gender_df.nunique()["SN"]

#Gender Percentage
gender_percent = (total_gender / player_count).map("{:,.2%}".format) 

#data frame
demographics_df = pd.DataFrame({"% of Players": gender_percent, "Players": total_gender})
demographics_df.sort_values(["Players"], ascending = False)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Drop duplicate Players
cleaned_players_df = purchase_data_df.drop_duplicates("SN")

#Create Age Bins
age_bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 99.99]
age_labels = ["< 10yrs", "10 -14yrs", "15-19yrs", "20-24yrs", "25-29yrs", "30-34yrs", "35-39yrs", "40yrs +"]

#Sort Purchases
purchase_data_df["Age Range"]= pd.cut(purchase_data_df['Age'],age_bins, labels= age_labels)

#Counts per Bin
total_age_purchase = cleaned_players_df["Age Range"].value_counts()

#percent of total
purchase_percent_age = (total_age_purchase / total_purchases).map("{:,.2%}".format)

#New Data Frame
age_purchase_demo_df = pd.DataFrame ({"Total Players":total_age_purchase , "Percentage of Players": purchase_percent_age})

#Show Data Frame
age_purchase_demo_df.reindex(["< 10yrs", "10 -14yrs", "15-19yrs", "20-24yrs", "25-29yrs", "30-34yrs", "35-39yrs", "40yrs +"])
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Spend Data
spend_data = purchase_data_df.groupby("SN")

#Player Purchase Count
player_purchase_count = spend_data["SN"].count().astype(int)

#Player Value
player_value = spend_data["Price"].sum()

#Average Purchase (Player)
player_avg_purchase = spend_data["Price"].mean().map("${:,.2f}".format) 

#Create New Data Frame
top_spenders_df = pd.DataFrame({"Purchase Count":player_purchase_count, 
                                "Average Purchase Price": player_avg_purchase, 
                                "Total Purchase Value": player_value})

#Format Total Value
top_spenders_df["Total Purchase Value"]= top_spenders_df["Total Purchase Value"].map("${:,.2f}".format) 

#Sort Data
top_spend_df = top_spenders_df.sort_values(["Purchase Count"], ascending=False).head(5)
top_spend_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results
|

* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python


# Group the item data by item id and item name 
popular_df = purchase_data_df.groupby(["Item ID","Item Name","Price"])

# Count the number of times an item has been purchased 
purchase_count = (popular_df["Price"].count()).astype(int)

# Calcualte the purchase value per item 
purchase_value = (popular_df["Price"].sum()) 

#Create Data Frame
popular_items_df = pd.DataFrame({"Purchase Count": purchase_count, "Total Purchase Value":purchase_value})

#Format Item Price
popular_items_df["Total Purchase Value"]= popular_items_df["Total Purchase Value"].map("${:,.2f}".format) 

#Sort and Display Data frame
most_popular_df = popular_items_df.sort_values(["Purchase Count"], ascending=False).head()
most_popular_df
```
